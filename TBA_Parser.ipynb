```python
kalimat = input('Kalimat yang ingin dibentuk :')
tokens = kalimat.lower().split()
tokens.append('EOS')

non_terminals = ['S','NN','VB']
terminals = ['pere','mere','chat','chapeau','utiliser','faire','voir','lunettes','pain','fleur']

parse_table = {}

parse_table[('S','pere')] = ['NN','VB','NN']
parse_table[('S','mere')] = ['NN','VB','NN']
parse_table[('S','chat')] = ['NN','VB','NN']
parse_table[('S','chapeau')] = ['NN','VB','NN']
parse_table[('S','lunettes')] = ['NN','VB','NN']
parse_table[('S','pain')] = ['NN','VB','NN']
parse_table[('S','fleur')] = ['NN','VB','NN']
parse_table[('S','utiliser')] = ['error']
parse_table[('S','faire')] = ['error']
parse_table[('S','voir')] = ['error']
parse_table[('S','EOS')] = ['error']

parse_table[('NN','pere')] = ['pere']
parse_table[('NN','mere')] = ['mere']
parse_table[('NN','chat')] = ['chat']
parse_table[('NN','chapeau')] = ['chapeau']
parse_table[('NN','lunettes')] = ['lunettes']
parse_table[('NN','pain')] = ['pain']
parse_table[('NN','fleur')] = ['fleur']
parse_table[('NN','utiliser')] = ['error']
parse_table[('NN','faire')] = ['error']
parse_table[('NN','voir')] = ['error']
parse_table[('NN','EOS')] = ['error']

parse_table[('VB','pere')] = ['error']
parse_table[('VB','mere')] = ['error']
parse_table[('VB','chat')] = ['error']
parse_table[('VB','chapeau')] = ['error']
parse_table[('VB','lunettes')] = ['error']
parse_table[('VB','pain')] = ['error']
parse_table[('VB','fleur')] = ['error']
parse_table[('VB','utiliser')] = ['utiliser']
parse_table[('VB','faire')] = ['faire']
parse_table[('VB','voir')] = ['voir']
parse_table[('VB','EOS')] = ['error']

stack = []
stack.append('#')
stack.append('S')

idx_token = 0
symbol = tokens[idx_token]

while (len(stack) > 0):
  top = stack[len(stack)-1]
  print('top = ',top)
  print('symbol = ',symbol)
  if top  in terminals:
    print('top adalah simbol terminal')
    if top==symbol:
      stack.pop()
      idx_token = idx_token + 1
      symbol = tokens[idx_token]
      if symbol == 'EOS':
        print('isi stack: ', stack)
        stack.pop()
    else:
      #comment
      print('error')
      break;
  elif top in non_terminals:
    print('top adalah simbol non-terminal')
    if parse_table[(top, symbol)][0] != 'error':
      stack.pop()
      symbols_to_be_pushed = parse_table[(top, symbol)]
      for i in range(len(symbols_to_be_pushed)-1,-1,-1):
        stack.append(symbols_to_be_pushed[i])
    else:
      print('error')
      break;
  else:
    print('error')
    break;
  print('isi stack:',stack)
  print()

print()
if symbol == 'EOS' and len(stack)==0:
  print('input string ', kalimat ,' diterima, sesuai grammar')
else:
  print('Error, input string:', kalimat, ', tidak diterima, tidak sesuai grammar')
```
