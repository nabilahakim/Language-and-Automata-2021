```python
import string

sentence = input("Masukkan kalimat/kata yang ingin diperiksa: ")
input_string = sentence.lower() + "#"

alphabet_list = list(string.ascii_lowercase)
state_list = ["q0", "q1","q2","q3","q4","q5","q6","q7","q8","q9","q10","q11","q12","q13","q14","q15","q16","q17","q18","q19","q20","q21","q22","q23","q24","q25","q26","q27","q28","q30","q31","q32","q33"]

transition_table = {}

for state in state_list:
	for alphabet in alphabet_list:
		transition_table[(state, alphabet)]= 'error'
	transition_table[(state, '#')] = 'error'
	transition_table[(state, ' ')] = 'error'

#before input string
transition_table['q0', ' '] = 'q0'

#finish_state
transition_table[("q8", "#")] = "accept"
transition_table[("q8", " ")] = "q9"

transition_table[("q9", "#")] = "accept"
transition_table[("q9", " ")] = "q9"


#input pere
transition_table[('q9', 'p')] = 'q12'
transition_table[('q0', 'p')] = 'q12'
transition_table[('q12', 'e')] = 'q13'
transition_table[('q13', 'r')] = 'q14'
transition_table[('q14', 'e')] = 'q8'

#input mere
transition_table[('q9', 'm')] = 'q12'
transition_table[('q0', 'm')] = 'q12'
transition_table[('q12', 'e')] = 'q13'
transition_table[('q13', 'r')] = 'q14'
transition_table[('q14', 'e')] = 'q8'

#input Chat
transition_table[('q9', 'c')] = 'q21'
transition_table[('q0', 'c')] = 'q21'
transition_table[('q21', 'h')] = 'q22'
transition_table[('q22', 'a')] = 'q23'
transition_table[('q23', 't')] = 'q8'

#Chapeau
transition_table[('q9', 'c')] = 'q21'
transition_table[('q0', 'c')] = 'q21'
transition_table[('q21', 'h')] = 'q22'
transition_table[('q22', 'a')] = 'q23'
transition_table[('q23', 'p')] = 'q24'
transition_table[('q24', 'e')] = 'q25'
transition_table[('q25', 'a')] = 'q26'
transition_table[('q26', 'u')] = 'q8'

#Utiliser
transition_table[('q9', 'u')] = 'q1'
transition_table[('q0', 'u')] = 'q1'
transition_table[('q1', 't')] = 'q2'
transition_table[('q2', 'i')] = 'q3'
transition_table[('q3', 'l')] = 'q4'
transition_table[('q4', 'i')] = 'q5'
transition_table[('q5', 's')] = 'q6'
transition_table[('q6', 'e')] = 'q7'
transition_table[('q7', 'r')] = 'q8'


#Faire
transition_table[('q9', 'f')] = 'q17'
transition_table[('q0', 'f')] = 'q17'
transition_table[('q17', 'a')] = 'q18'
transition_table[('q18', 'i')] = 'q19'
transition_table[('q19', 'r')] = 'q20'
transition_table[('q20', 'e')] = 'q8'

#Voir
transition_table[('q9', 'v')] = 'q10'
transition_table[('q0', 'v')] = 'q10'
transition_table[('q10', 'o')] = 'q11'
transition_table[('q11', 'i')] = 'q7'
transition_table[('q7', 'r')] = 'q8'

#Lunettes
transition_table[('q9', 'l')] = 'q27'
transition_table[('q0', 'l')] = 'q27'
transition_table[('q27', 'u')] = 'q28'
transition_table[('q28', 'n')] = 'q29'
transition_table[('q29', 'e')] = 'q30'
transition_table[('q30', 't')] = 'q31'
transition_table[('q31', 't')] = 'q32'
transition_table[('q32', 'e')] = 'q33'
transition_table[('q33', 's')] = 'q8'

#Pain
transition_table[('q9', 'p')] = 'q12'
transition_table[('q0', 'p')] = 'q12'
transition_table[('q12', 'a')] = 'q15'
transition_table[('q15', 'i')] = 'q16'
transition_table[('q16', 'n')] = 'q8'

#Fleur
transition_table[('q9', 'f')] = 'q17'
transition_table[('q0', 'f')] = 'q17'
transition_table[('q17', 'l')] = 'q12'
transition_table[('q12', 'e')] = 'q13'
transition_table[('q13', 'u')] = 'q19'
transition_table[('q19', 'r')] = 'q8'



idx_char = 0
state = "q0"
current_token = ""
while state != "accept":
    current_char = input_string[idx_char]
    current_token += current_char
    state = transition_table[(state, current_char)]
    if state == "q9":
        print("current token: ",current_token,',valid')
        current_token = ""
    if state == "error":
        print("error")
        break
    idx_char += 1

if state=='accept':
	print('semua token di input'  , sentence, ', valid')



```

```python
kalimat = input('Kalimat yang ingin dibentuk :')
tokens = kalimat.lower().split()
tokens.append('EOS')

non_terminals = ['S','NN','VB']
terminals = ['pere','mere','chat','chapeau','utiliser','faire','voir','lunettes','pain','fleur']

parse_table = {}

parse_table[('S','pere')] = ['NN','VB','NN']
parse_table[('S','mere')] = ['NN','VB','NN']
parse_table[('S','chat')] = ['NN','VB','NN']
parse_table[('S','chapeau')] = ['NN','VB','NN']
parse_table[('S','lunettes')] = ['NN','VB','NN']
parse_table[('S','pain')] = ['NN','VB','NN']
parse_table[('S','fleur')] = ['NN','VB','NN']
parse_table[('S','utiliser')] = ['error']
parse_table[('S','faire')] = ['error']
parse_table[('S','voir')] = ['error']
parse_table[('S','EOS')] = ['error']

parse_table[('NN','pere')] = ['pere']
parse_table[('NN','mere')] = ['mere']
parse_table[('NN','chat')] = ['chat']
parse_table[('NN','chapeau')] = ['chapeau']
parse_table[('NN','lunettes')] = ['lunettes']
parse_table[('NN','pain')] = ['pain']
parse_table[('NN','fleur')] = ['fleur']
parse_table[('NN','utiliser')] = ['error']
parse_table[('NN','faire')] = ['error']
parse_table[('NN','voir')] = ['error']
parse_table[('NN','EOS')] = ['error']

parse_table[('VB','pere')] = ['error']
parse_table[('VB','mere')] = ['error']
parse_table[('VB','chat')] = ['error']
parse_table[('VB','chapeau')] = ['error']
parse_table[('VB','lunettes')] = ['error']
parse_table[('VB','pain')] = ['error']
parse_table[('VB','fleur')] = ['error']
parse_table[('VB','utiliser')] = ['utiliser']
parse_table[('VB','faire')] = ['faire']
parse_table[('VB','voir')] = ['voir']
parse_table[('VB','EOS')] = ['error']

stack = []
stack.append('#')
stack.append('S')

idx_token = 0
symbol = tokens[idx_token]

while (len(stack) > 0):
  top = stack[len(stack)-1]
  print('top = ',top)
  print('symbol = ',symbol)
  if top  in terminals:
    print('top adalah simbol terminal')
    if top==symbol:
      stack.pop()
      idx_token = idx_token + 1
      symbol = tokens[idx_token]
      if symbol == 'EOS':
        print('isi stack: ', stack)
        stack.pop()
    else:
      #comment
      print('error')
      break;
  elif top in non_terminals:
    print('top adalah simbol non-terminal')
    if parse_table[(top, symbol)][0] != 'error':
      stack.pop()
      symbols_to_be_pushed = parse_table[(top, symbol)]
      for i in range(len(symbols_to_be_pushed)-1,-1,-1):
        stack.append(symbols_to_be_pushed[i])
    else:
      print('error')
      break;
  else:
    print('error')
    break;
  print('isi stack:',stack)
  print()

print()
if symbol == 'EOS' and len(stack)==0:
  print('input string ', kalimat ,' diterima, sesuai grammar')
else:
  print('Error, input string:', kalimat, ', tidak diterima, tidak sesuai grammar')
```
